```python
from cdl_helper import head_to_head, read_in_all_matches

import matplotlib.pyplot as plt
import pandas as pd

df = read_in_all_matches()
df.sample(5)
```

I will remove the players who are not on the current starting rosters

```python
df = df[~df['alias'].isin(['Prolute', 'Capsidal'])]
```

```python
fig = plt.figure(figsize=(14, 18))
((ax3, ax4),
 (ax5, ax6),
 (ax7, ax8)) = fig.subplots(ncols=2, nrows=3)

head_to_head(df, 'LV', 'LAT', 'kd', "Vegas Legion kd vs Los Angeles Thieves compared to vs rest of league",ax3, gamemode='CDL Hardpoint')
head_to_head(df, 'LAT', 'LV', 'kd', "Los Angeles Thieves kd vs Vegas Legion compared to vs rest of league",ax4, gamemode='CDL Hardpoint')
head_to_head(df, 'LV', 'LAT', 'totalDamageDealt', "Vegas Legion damage vs Los Angeles Thieves compared to vs rest of league",ax5, gamemode='CDL Hardpoint')
head_to_head(df, 'LAT', 'LV', 'totalDamageDealt', "Los Angeles Thieves damage vs Vegas Legion compared to vs rest of league",ax6, gamemode='CDL Hardpoint')
head_to_head(df, 'LV', 'LAT', 'accuracy', "Vegas Legion accuracy vs Los Angeles Thieves compared to vs rest of league",ax7, gamemode='CDL Hardpoint')
head_to_head(df, 'LAT', 'LV', 'accuracy', "Los Angeles Thieves accuracy vs Vegas Legion compared to vs rest of league",ax8, gamemode='CDL Hardpoint')

fig.suptitle("Vegas Legion And Los Angeles Thieves Head to Head Performances (Hardpoint)");
```
